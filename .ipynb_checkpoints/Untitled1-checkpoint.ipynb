```python
import torch
import gc
import matplotlib.pyplot as plt

from torch import nn, optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from jcopdl.layers import linear_block
from jcopdl.callback import Callback, set_config
from tqdm.auto import tqdm

device=torch.device("cuda" if torch.cuda.is_available() else "cpu")
gc.collect()
torch.cuda.empty_cache()
device
```

```python
bs=512
crop_size=64

train_transform=transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(45),
    transforms.RandomResizedCrop(crop_size, scale=(0.8,1.0)),
    transforms.ToTensor()
])

test_transform=transforms.Compose([
    transforms.Resize((crop_size+10)),
    transforms.CenterCrop(crop_size),
    transforms.ToTensor()
])

train_dir='Dataset/Train'
train_set=datasets.ImageFolder(train_dir, transform=train_transform)
trainloader=DataLoader(train_set, batch_size=bs, shuffle=True, num_workers=2)

test_dir='Dataset/Validation'
test_set=datasets.ImageFolder(test_dir, transform=test_transform)
testloader=DataLoader(test_set, batch_size=bs, shuffle=True, num_workers=2) 
```

```python
f,t=next(iter(trainloader))
f.shape
```

```python
from jcopdl.layers import conv_block,linear_block
```

```python
class CustomClassifier(nn.Module):
    def __init__(self, crop_size, n1, n2, n3, n4, output_size, dropout=0.2, n_filter=3, n_stride=1, n_padding=1):
        super().__init__()
        self.conv=nn.Sequential(
            conv_block(3,n1),
            conv_block(n1,n2),
            conv_block(n2,n3),
            conv_block(n3,n4),
            
            nn.Flatten()
        )
        
        
        self.fc=nn.Sequential(
            
            linear_block(64*4*4, 512, dropout=dropout),
            linear_block(512, 256, dropout=dropout),
            linear_block(256, 64, dropout=dropout),
            linear_block(64, output_size, activation='lsoftmax')
        )
        
    def forward(self, x):
        x = self.conv(x)
        x = self.fc(x)
        return x
```

```python
config= set_config({
    "crop_size":crop_size, 
    "n_filter":3, 
    "n_stride":1, 
    "n_padding":1,
    "batch_size":bs,
    "n1":8,
    "n2":16,
    "n3":32,
    "n4":64,
    "output_size": len(train_set.classes),
    "dropout":0.4
})
```

```python
model=CustomClassifier(config.crop_size, config.n1, config.n2, config.n3, config.n4, config.output_size, config.dropout, config.n_filter, config.n_stride, config.n_padding).to(device)
citerion=nn.NLLLoss()
optimizer=optim.AdamW(model.parameters(), lr=0.001)
callback=Callback(model, config, outdir="model", save_every=20, early_stop_patience=10, plot_every= 20)
```

```python
model.parameters
```

```python
def loop_fn(mode, dataset, dataloader, model, citerion, optimizer, device):
    if mode =='train':
        model.train()
    elif mode =='test':
        model.eval()
    cost = correct = 0
    for feature, target in tqdm(dataloader, desc=mode.title()):
        feature, target=feature.to(device), target.to(device)
        output = model(feature)
        loss = citerion(output, target)
        
        if mode == 'train':
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
        
        cost += loss.item() * feature.shape[0]
        correct += (output.argmax(1) == target).sum().item()
    cost = cost / len(dataset)
    acc = correct / len(dataset)
    return cost, acc    
```

```python
while True:
    train_cost, train_score = loop_fn("train", train_set, trainloader, model, citerion, optimizer, device)
    with torch.no_grad():
        test_cost, test_score = loop_fn("test", test_set, testloader, model, citerion, optimizer, device)
        
    callback.log(train_cost, test_cost, train_score, test_score)
    callback.save_checkpoint()
    callback.cost_runtime_plotting()
    callback.score_runtime_plotting()
    
    if callback.early_stopping(model, monitor="test_score"):
        callback.plot_cost()
        callback.plot_score()
        break
```

```python
newdata_dir='newdata'
newdata_set=datasets.ImageFolder(newdata_dir, transform=test_transform)
newdataloader=DataLoader(newdata_set, batch_size=bs, shuffle=True, num_workers=2)

feature, target = next(iter(newdataloader))
feature, target = feature.to(device), target.to(device)

label2cat=train_set.classes

with torch.no_grad():
    model.eval()
    output = model(feature)
    predictions = output.argmax(1)
```

```python
fig, axes = plt.subplots(5, 5, figsize=(25, 25))
for img, label, prediction, ax in zip(feature, target, predictions, axes.flatten()):
    ax.imshow(img.permute(1,2,0).cpu())
    font= {'color' : 'r'} if label != prediction else {'color' : 'g'}
    label, prediction = label2cat[label.item()], label2cat[prediction.item()]
    ax.set_title(f" Label: {label} | Pred: {prediction}",fontdict = font);
    ax.axis('off')
```
